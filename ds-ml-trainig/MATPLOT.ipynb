```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x1 = [0, 1, 2, 3, 4, 5, 6, 7, 9, 10]
y1 = [1, 2, 5, 10, 17, 26, 37, 50, 82, 101]

x2 = np.arange(0, 40, 2)
y2 = 2 * x2

# Plotting the graphs
plt.plot(x1, y1, label='Graph 1')
plt.plot(x2, y2, label='Graph 2')

# Adding labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

#plt.axis([0,10,10,60])

# Finding x when y is 60 for both plots
x1_at_60 = None
x2_at_60 = None

for i in range(len(y1)):
    if y1[i] == 60:
        x1_at_60 = x1[i]

for i in range(len(y2)):
    if y2[i] == 60:
        x2_at_60 = x2[i]

# Printing the outputs
print(x1_at_60)
print(x2_at_60)

# Displaying the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the first plot
x1 = [0, 1, 2, 3, 4, 5, 6, 7, 9, 10]
y1 = [1, 2, 5, 10, 17, 26, 37, 50, 82, 101]

# Data for the second plot
x2 = np.arange(0, 40, 2)
y2 = 2 * x2

# Plotting the graphs
plt.plot(x1, y1, label='Graph 1')
plt.plot(x2, y2, label='Graph 2')

# Setting labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Finding x when y is 60 for both plots
x1_y60 = [x for x, y in zip(x1, y1) if y == 60]
x2_y60 = [x for x, y in zip(x2, y2) if y == 60]

# Printing the output
print(min(x1_y60), max(x1_y60))
print(min(x2_y60), max(x2_y60))

# Displaying the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
company = ['HP','Dell','Lenovo','Asus','Apple','Acer']
sold = [20000,43000,15000,17000,22000,13000]

company = np.array(company)
sold = np.array(sold)

plt.pie(sold, labels=company, autopct='%1.1f%%', counterclock=False)
plt.axis('equal')
plt.show()

x = np.true_divide(sold, sold.sum(axis=0, keepdims=True))*100
for i in range(len(company)):
    print(company[i],format(x[i],'.1f'))
```
